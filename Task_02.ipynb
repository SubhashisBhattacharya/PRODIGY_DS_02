```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<b><h2>Loading the dataset</h2></b>

```python
data= pd.read_csv('test.csv')
data.head(10)
```

<b><h2>Checking for missing values</b></h2>

```python
data.isnull().sum()
```

<b><h2>Dealing with missing values</b></h2>

```python
data['Age']=data['Age'].fillna(data['Age'].median())
```

```python
data.drop(columns=['Cabin'],inplace = True)
```

<b><h2>Checking for missing values</b></h2>

```python
data.isnull().sum()
```

<b><h2>Displaying the five point summary of the data</b></h2>

```python
data.describe()
```

<b><h2>Visualization of the data</b></h2>

```python
sns.histplot(data['Age'],bins=30,kde=True,color='red')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Freaquency')
plt.show() # Age distribution suggests that most passengers were bwtween ages 20 to 30
```

```python
numeric_data = data.select_dtypes(include=['number'])
plt.figure(figsize=(10,6))
sns.heatmap(numeric_data.corr(),annot=True,fmt='2f')
plt.title('Correlation Heatmap')
plt.show()
```
